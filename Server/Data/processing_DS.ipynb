### Processing the "Chess All GM players Statistics 2023" Dataset.

```python
import pandas as pd
import sklearn as sk
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('GM_players_statistics.csv')
print(df.shape)
df.head()
```

#### Removing Unnecessary Columns 

```python
df = df.drop(['highest_rapid_date','highest_blitz_date','highest_bullet_date','profile', 'country', 'joined', 'last_online', 'highest_tactics_date', 'is_streamer', 'status', 'league','title', 'Unnamed: 0', 'username'], axis=1)
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df = df.set_index('user_id') # set user_id as index
print(df.shape)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# normalizing the data
scaler = MinMaxScaler()
for column in df.columns:
    if column != 'name' and column != 'user_id':
        df[column] = scaler.fit_transform(df[[column]])
    
df.head()
```

```python
df.to_csv('GM_players_statistics_cleaned.csv')
```
